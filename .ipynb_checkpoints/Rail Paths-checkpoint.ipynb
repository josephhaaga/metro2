```python
import pandas as pd;
import numpy as np;
import json
key = open("keys.py").readline().replace("\n","")
```

```python
########### Python 3.2 #############
import http.client, urllib.request, urllib.parse, urllib.error, base64

headers = {
    # Request headers
    'api_key': key,
}

params = urllib.parse.urlencode({
})

try:
    conn = http.client.HTTPSConnection('api.wmata.com')
    conn.request("GET", "/Rail.svc/json/jLines?%s" % params, "{body}", headers)
    response = conn.getresponse()
    data = response.read().decode('utf-8')
    print(data)
    conn.close()
except Exception as e:
    print("[Errno {0}] {1}".format(e.errno, e.strerror))

####################################
```

```python
result = json.loads(data)
# print (json.dumps(result,indent=4))
```

```python
paths = {}
master_paths = pd.DataFrame();
for i in result['Lines']:
    params = urllib.parse.urlencode({
        # Request parameters
        'FromStationCode': i['StartStationCode'],
        'ToStationCode': i['EndStationCode'],
    })

    try:
        conn = http.client.HTTPSConnection('api.wmata.com')
        conn.request("GET", "/Rail.svc/json/jPath?%s" % params, "{body}", headers)
        response = conn.getresponse()
        data = response.read().decode('utf-8')
        print(data)
        conn.close()
    except Exception as e:
        print("[Errno {0}] {1}".format(e.errno, e.strerror))

    ####################################
    paths[i['DisplayName']] = pd.DataFrame(json.loads(data)['Path'])
    master_paths.append(json.loads(data)['Path'])
```

```python
station_names = []
paths.keys()
for line in paths.keys():
    for station in paths[line]['StationName']:
        station_names.append(station);
    
```

```python
def findLines(station):
    stations = [];
    for line in paths.keys():
        for stat in paths[line]['StationName']:
            if stat==station:
                stat_color = paths[line][paths[line]['StationName']==stat]['LineCode'];
                index = str(stat_color).index("    ")+4;
#                 print (str(stat_color)[index:index+2])
                stations.append((str(stat_color)[index:index+2]))
    return stations;
```

```python
shortcode_map = {"RD":"Red", "YL":"Yellow", "BL":"Blue", "GR":"Green", "SV":"Silver", "OR":"Orange"}
```

```python
adj_matrix = {};

for n in station_names:
    current_station = {};
    for k in findLines(n):
        for z in paths[shortcode_map[k]]['StationName']:
    #         Get distance from n to z via line k
            indexOfZ = (paths[shortcode_map[k]][paths[shortcode_map[k]]['StationName']==z].index.tolist()[0])
            indexOfN = (paths[shortcode_map[k]][paths[shortcode_map[k]]['StationName']==n].index.tolist()[0])
            current_station[z]=(k,int(abs(indexOfZ - indexOfN)));
    adj_matrix[n] = current_station
```

```python
def pathsToDestination(origin,dest):
    tempPaths = {};
    for n in findLines(origin):
        this_line = paths[shortcode_map[n]];
        if(len(this_line[this_line['StationName']==destination])>0):
            try:
                indexOfZ = (paths[shortcode_map[n]][paths[shortcode_map[n]]['StationName']==origin].index.tolist()[0])
                indexOfN = (paths[shortcode_map[n]][paths[shortcode_map[n]]['StationName']==dest].index.tolist()[0])
                tempPaths[n]=int(abs(indexOfZ - indexOfN));
            except:
                pass;
    return tempPaths;
```

```python
lenfantToBraddock = pathsToDestination("L'Enfant Plaza","Braddock Road")
```

```python
master_adj_matrix = {};
for n in station_names:
    current_station = {};
    for g in station_names:
        if(len(pathsToDestination(n,g))>0):
            current_station[g]=pathsToDestination(n,g);
    master_adj_matrix[n] = current_station
```

```python
def getSingleLinePath(start,destination):
    possible_paths = {}
    originLines = findLines(start);
    destLines = findLines(destination);
    commonLines = list(set(originLines).intersection(destLines))
    for c in commonLines:
        color=shortcode_map[c]
        print (color)
        st_index = list(paths[color][paths[color]['StationName']==start].index)[0]
        dest_index = list(paths[color][paths[color]['StationName']==destination].index)[0]
        if(dest_index>st_index):
            single_line_path = paths[color].iloc[st_index:dest_index];
        else:
            single_line_path = paths[color].iloc[dest_index:st_index];
        possible_paths[c] = list(single_line_path['StationName']);
    return possible_paths
```
