```python
########### Python 3.2 #############
import http.client, urllib.request, urllib.parse, urllib.error, base64
import pandas as pd

key = open("keys.py").readline().replace("\n","")

# stationCode="J03"
stationCode="All"

headers = {
    # Request headers
    'api_key': key,
}

params = urllib.parse.urlencode({
})

try:
    conn = http.client.HTTPSConnection('api.wmata.com')
    conn.request("GET", "/StationPrediction.svc/json/GetPrediction/"+stationCode+"?%s" % params, "{body}", headers)
    response = conn.getresponse()
    data = response.read().decode('utf-8')
    print(data)
    conn.close()
except Exception as e:
    print("[Errno {0}] {1}".format(e.errno, e.strerror))

####################################
```

```python
import json
z = pd.DataFrame(json.loads(data)['Trains'])

# Construct dict of location codes and names
stations = {};
for x in z.LocationCode.unique():
    stations[x] = z[z['LocationCode']==x].iloc[0]['LocationName'];
```

```python
# If I am at U St heading to Braddock Rd, is it faster to wait for Yellow line?
# or take Green to L'Enfant and transfer to Blue line?
U_street_trains = z[z['LocationName']=="U Street/African-Amer Civil War Memorial/Cardozo"]
# Yellow line to Franconia-Springfield or Huntington will get me to Braddock Rd
huntington_bound = U_street_trains[U_street_trains['DestinationName']=='Huntington']
franconia_bound = U_street_trains[U_street_trains['DestinationName']=='Franconia-Springfield']
```

```python
############## Travel Time from USt to Huntington or Franconia ################ 
params = urllib.parse.urlencode({
    # Request parameters
    'FromStationCode': 'E03',
    'ToStationCode': df_new['DestinationCode'].iloc[0],
})

try:
    conn = http.client.HTTPSConnection('api.wmata.com')
    conn.request("GET", "/Rail.svc/json/jSrcStationToDstStationInfo?%s" % params, "{body}", headers)
    response = conn.getresponse()
    data = response.read()
    print(data)
    conn.close()
except Exception as e:
    print("[Errno {0}] {1}".format(e.errno, e.strerror))

####################################
```
